```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('HR_Attrition.csv')
```

```python
df.head()
```

# 1. Subplots:

### Generate a figure with 2x2 subplots, each containing a different type of plot (e.g., line plot, scatter plot, bar plot, and histogram).

```python
plt.subplot(2, 2, 1)
plt.plot(df['BusinessTravel'].unique(), df['Department'].value_counts())

plt.subplot(2, 2, 2)
plt.scatter(df['YearsInCurrentRole'], df['PercentSalaryHike'])

plt.subplot(2, 2, 3)
plt.bar(df['Age'].unique(), df.groupby('Age')['MonthlyIncome'].mean())

plt.subplot(2, 2, 4)
plt.hist(df['DistanceFromHome'])
```

```python
df.columns
```

### Customize the subplots with titles, axis labels, and legends.

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

axs[0, 0].plot(df['BusinessTravel'].unique(), df['Department'].value_counts(), color='green', marker='o', markersize=5, linestyle='dashed')
axs[0, 0].set_title('Business Travel by Department')
axs[0, 0].set_xlabel('Business Travel')
axs[0, 0].set_ylabel('Department')
axs[0, 0].legend(['Department'])

axs[0, 1].scatter(df['PercentSalaryHike'], df['YearsInCurrentRole'], cmap='Spectral', alpha=0.75, c = (np.random.rand(1470) * 10), s=(np.pi * (5 * np.random.rand(1470))**2))
axs[0, 1].set_title('% Hike by Years in Current Role')
axs[0, 1].set_xlabel('% Hike')
axs[0, 1].set_ylabel('Years')

axs[1, 0].bar(df['Age'].unique(), df.groupby('Age')['MonthlyIncome'].mean(), color='cyan', edgecolor='green')
axs[1, 0].set_title('Average Monthly Income by Age')
axs[1, 0].set_xlabel('Age')
axs[1, 0].set_ylabel('Average Income')
axs[1, 0].legend(['Average Income'])

axs[1, 1].hist(df['DistanceFromHome'], bins=25, color='lightcoral', edgecolor='black')
axs[1, 1].set_title('Distance From Home')
axs[1, 1].set_xlabel('Distance')
axs[1, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

# 2. Histograms:

### Create a histogram for a given numerical data set. Customize the number of bins.

```python
plt.hist(df['HourlyRate'], bins=20, color='grey', edgecolor='g', label='Hourly Rate')
plt.title('Hourly Rate')
plt.xlabel('Value')
plt.ylabel('Count')
plt.legend()

plt.show()
```

### Plot multiple histograms on the same figure to compare different data sets.

```python
plt.hist(df['TotalWorkingYears'], bins=20, color='skyblue', edgecolor='g')

plt.hist(df['YearsAtCompany'], bins=20, color='lightcoral', edgecolor='b')

plt.title('Total Working Years / Years at Company')
plt.xlabel('Years')
plt.ylabel('Count')
plt.legend(df[['TotalWorkingYears', 'YearsAtCompany']])

plt.show()
```

### Normalize the histograms and add appropriate labels and titles.

```python
plt.hist(df['TotalWorkingYears'], bins=20, color='skyblue', edgecolor='g', density=True)

plt.hist(df['YearsAtCompany'], bins=20, color='lightcoral', edgecolor='b', density=True)

plt.title('Total Working Years / Years at Company')
plt.xlabel('Years')
plt.ylabel('Count')
plt.legend(df[['TotalWorkingYears', 'YearsAtCompany']])

plt.show()
```

# 3. Box Plots:

### Create a box plot for a given numerical data set.

```python
plt.boxplot(df['MonthlyIncome'])
plt.ylabel('Monthly Income')
plt.title('Box Plot of the Monthly Income')

plt.show()
```

```python
df.columns
```

### Generate multiple box plots in a single figure to compare different data sets.

```python
plt.boxplot([df['YearsInCurrentRole'], df['YearsAtCompany'], df['YearsWithCurrManager'], df['YearsSinceLastPromotion']], labels=['Current Role', 'Company', 'Current Manager', 'Last Promotion'])
plt.ylabel('Years')
plt.title('Box Plot of Years')

plt.show()
```

### Customize the box plots with titles, labels, and change the colors.

```python
plt.boxplot([df['YearsInCurrentRole'], df['YearsAtCompany'], df['YearsWithCurrManager'], df['YearsSinceLastPromotion']], 
            labels=['Current Role', 'Company', 'Current Manager', 'Last Promotion'], patch_artist=True, 
            boxprops= dict(facecolor='lightcoral', color='green'), notch=True, medianprops= dict(color='blue'),
           whiskerprops= dict(color='green'), capprops= dict(color='m'), flierprops= dict(marker='+', markersize=5))
plt.ylabel('Years')
plt.title('Box Plot of Years')

plt.show()
```
