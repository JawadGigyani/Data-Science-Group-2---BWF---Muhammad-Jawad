```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## 1. Linear Regression Plot (sns.regplot)

```python
sns.regplot(x='total_bill', y='tip', data=tips)
plt.title('Linear Regression Plot')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## 2. Multiple Linear Regression Plot (sns.lmplot)

```python
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.title('Multiple Linear Regression Plot')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## 3. Residual Plot (sns.residplot)

```python
sns.residplot(x='total_bill', y='tip', lowess=True, data=tips)
plt.title('Residual Plot')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## 4. Histogram (sns.histplot)


```python
sns.histplot(tips['total_bill'], bins=30, color='g', edgecolor='c')
plt.title('Histogram')
plt.xlabel('Total Bill')
plt.show()
```

## 5. Kernel Density Estimate (KDE) Plot (sns.kdeplot)


```python
sns.kdeplot(tips['total_bill'], shade=True, color='g')
plt.title('Kernel Density Estimate Plot')
plt.xlabel('Total Bill')
plt.show()
```

### Bivariate Kdeplot

```python
sns.kdeplot(x='total_bill', y='tip',shade=True, color='c', cmap='coolwarm', cbar=True,data=tips)
plt.title('Bivariate Kernel Density Estimate Plot')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## 6. Distribution Plot (sns.distplot)
* According to the Seaborn Documentation `distplot()` is deprecated and has been replaced by `histplot()` and `displot()`.

```python
sns.histplot(tips['total_bill'], kde=True, color='brown')
plt.title('Distribution Plot')
plt.xlabel('Total Bill')
plt.show()
```

```python
sns.displot(tips['total_bill'], kde=True, color='lightcoral')
plt.title('Distribution Plot')
plt.xlabel('Total Bill')
plt.show()
```

## 7. Box Plot (sns.boxplot)


```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

## 8. Violin Plot (sns.violinplot)


```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

## Load `iris` Dataset

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## 9. Pair Plot (sns.pairplot)

```python
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.title('Pair Plot')
plt.show()
```

```python
sns.pairplot(iris, hue='species', kind='hist')
plt.title('Pair Plot')
plt.show()
```

```python
sns.pairplot(iris, hue='species', diag_kind='hist')
plt.title('Pair Plot')
plt.show()
```

## Load `flights` Dataset

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

## 10. Heatmap (sns.heatmap)


```python
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='coolwarm', linewidths=1, linecolor='g')
plt.title('Heatmap')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

## 11. Joint Plot (sns.jointplot)

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', color='g')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

## 12. Facet Grid (sns.FacetGrid)

```python
g = sns.FacetGrid(tips, col='sex', row='smoker')
g.map(sns.histplot, 'total_bill')
g.fig.suptitle('Facet Grid', y=1.02)
plt.show()
```

```python
g = sns.FacetGrid(tips, col='sex', row='smoker', hue='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')
g.fig.suptitle('Facet Grid', y=1.02)
plt.show()
```
