```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('Service_Usage.csv')
df2 = pd.read_csv('Subscriptions.csv')
df3 = pd.read_csv('Customer.csv')
```

```python
df1.head()
```

```python
df2.head()
```

## 1. Merge two DataFrames on a single key.

```python
merged_df_single_key = pd.merge(df1, df2, on='usage_id')
```

```python
merged_df_single_key.head()
```

## 2. Merge two DataFrames on multiple keys.

```python
merged_df_multiple_keys = pd.merge(df1, df2, on=['usage_id', 'customer_id'])
```

```python
merged_df_multiple_keys.head()
```

## 3. Perform an outer join, inner join, left join, and right join.

#### Outer Join

```python
outer_join_df = pd.merge(df1, df2, on='customer_id', how='outer')
```

```python
outer_join_df.head()
```

#### Inner Join

```python
inner_join_df = pd.merge(df1, df2, on='customer_id', how='inner')
```

```python
inner_join_df.head()
```

#### Left Join

```python
left_join_df = pd.merge(df1, df2, on='customer_id', how='left')
```

```python
left_join_df.head()
```

#### Right Join

```python
right_join_df = pd.merge(df1, df2, on='customer_id', how='right')
```

```python
right_join_df.head()
```

## 4. Concatenate two DataFrames along rows.

```python
concat_rows_df = pd.concat([df1, df2], axis=0)
```

```python
concat_rows_df.head()
```

## 5. Concatenate two DataFrames along columns.

```python
concat_cols_df = pd.concat([df1, df2], axis=1)
```

```python
concat_cols_df.head()
```

## 6. Concatenate a list of DataFrames.

```python
concat_list_df = pd.concat([df1, df2, df3], axis=1)
```

```python
concat_list_df.head()
```

## 7. Reshape data using the `melt` function to go from wide to long format.

```python
melt_df = pd.melt(df1, id_vars=['usage_id', 'customer_id', 'service_type', 'usage_date'], 
                    value_vars=['data_used', 'minutes_used'], 
                    var_name='usage_metric', value_name='value')
```

```python
melt_df.head()
```

## 8. Create a pivot table to summarize data.

```python
pivot_table = pd.pivot_table(df1, values=['data_used', 'minutes_used'], index='service_type', aggfunc={'data_used': 'sum', 'minutes_used': 'sum'})
```

```python
pivot_table
```

## 9. Group data by one or more columns and perform aggregation functions (e.g., sum, mean, count).

#### Group by one Column

```python
grouped_data_one = df1.groupby(['service_type']).agg({'data_used': ['mean'], 'minutes_used': ['sum']})
```

```python
grouped_data_one
```

#### Group by multiple Column

```python
grouped_data_mul = df1.groupby(['service_type', 'usage_date']).agg({'customer_id': ['count'] ,'minutes_used': ['mean']})
```

```python
grouped_data_mul.head()
```

## 10. Apply multiple aggregation functions to grouped data.

```python
mul_agg = df1.groupby(['service_type']).agg({'customer_id': ['count'] ,'minutes_used': ['mean', 'mean'], 'data_used': ['sum', 'std']})
```

```python
mul_agg
```

## 11. Use the groupby function to group data and apply custom functions.

```python
def custom_func(x):
    return x.max() - x.min()
```

```python
df1.groupby('customer_id')[['data_used', 'minutes_used']].apply(custom_func).head()
```
