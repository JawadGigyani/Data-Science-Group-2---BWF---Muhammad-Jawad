## 8. Group the DataFrame by a specific column and calculate the mean of each group.

```python
group_col = df.groupby(['Gender']).mean(numeric_only=True)
group_col
```

## 9. Group the DataFrame by multiple columns and calculate the sum of each group.

```python
group_col_multiple = df.groupby(['City', 'Customer type','Product line']).sum(numeric_only=True)
group_col_multiple
```

## 10. Use the agg method to apply multiple aggregation functions to grouped data.

```python
group_agg = df.groupby(['City']).agg({'Branch': ['count'], 'Unit price': ['mean', 'std'], 'Quantity': ['median', 'nunique'], 'Total': ['mean'], 'Rating': ['mean', 'median']})
group_agg
```

## 11. Calculate the size of each group.

```python
print(group_col.size)
print(group_col_multiple.size)
print(group_agg.size)
```

## 12. Select rows based on multiple conditions.

```python
df[(df['City'] == 'Yangon') & (df['Customer type'] == 'Member') & (df['Total'] > 500) & (df['Payment'] == 'Credit card')]
```

## 13. Use the query method to filter rows.

```python
df.query('City == "Yangon" and `Unit price` > 80 and Rating > 9')
```

## 14. Use isin to filter rows based on a list of values.

```python
df[(df['Branch'].isin(['A', 'B'])) & (df['Product line'].isin(['Sports and travel', 'Fashion accessories']))]
```
