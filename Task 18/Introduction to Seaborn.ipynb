```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('HR_Attrition.csv')
```

```python
df.head()
```

# Scatter Plot

```python
sns.set_style('dark')

sns.scatterplot(
    data=df, 
    x='MonthlyIncome', 
    y='TotalWorkingYears', 
    palette='hls',
    hue='Gender', 
    size='Gender'
)

plt.xlabel('Monthly Salary')
plt.ylabel('Total Working Hour')
plt.title('Monthly Salary by Working Hours')

plt.show()
```

# Line Plot

```python
sns.set_style('whitegrid')

sns.lineplot(
    data=df, 
    x='NumCompaniesWorked', 
    y='PercentSalaryHike', 
    marker='o',
    palette='hls',
    hue='Education',
    ci=None
)

plt.xlabel('Years Worked at Compnay')
plt.ylabel('Salary Hike %')
plt.title('Salary Hike by Years Worked at Company')

plt.show()
```

# Bar Plot

```python
sns.set_style('darkgrid')

plt.figure(figsize=(15, 7))
sns.barplot(
    data=df,
    x='HourlyRate',
    y='TotalWorkingYears',
    palette='Set2',
)

plt.xlabel('Hourly Rate')
plt.ylabel('Total Working Years')
plt.title('Hourly Rate by Total Working Years')

plt.show()
```

# Count Plot

```python
sns.set_style('whitegrid')

sns.countplot(
    data=df,
    x='YearsSinceLastPromotion',
    palette='Paired'
    )

plt.xlabel('Years Since Last Promotion')
plt.ylabel('Count of Employees')
plt.title('Number of Employees by Years Since last Promotion')

plt.show()
```

# Box Plot

```python
sns.set_style('darkgrid')

sns.boxplot(
    data=df,
    x='DistanceFromHome',
    y='Attrition',
    palette='tab10',
    hue='EducationField',
    )

plt.xlabel('Distance From Home')
plt.ylabel('Attrition')
plt.title('Attrtion by Distance From Home')

plt.show()
```
