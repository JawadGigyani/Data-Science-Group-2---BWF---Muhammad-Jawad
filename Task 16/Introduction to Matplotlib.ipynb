```python
import matplotlib.pyplot as plt
import numpy as np
```

# 1. Basic Plots:

### Create a simple line plot.

```python
plt.plot([1,4,8], [1,4,8])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

### Plot multiple lines on the same graph.

```python
plt.title('Multiple Lines Plot')

plt.plot(np.arange(-10, 11), (x ** 2))

plt.plot([0, 0], [-10, 100])

plt.xlabel('X')
plt.ylabel('X Squared')
plt.show()
```

### Add markers to a line plot.

```python
x = np.arange(-10, 11)
print(x)

y = (x ** 2)
print(y)
```

```python
plt.title("Line with Marker")

plt.plot(x, x + 0, marker='o', markersize=5)
plt.plot(x, x + 1, marker='P', markersize=5)
plt.plot(x, x + 2, marker='s', markersize=5)
plt.plot(x, x + 3, marker='2', markersize=5)
```

### Customize line styles and colors.

```python
x1 = np.linspace(0, 2*np.pi, 200)
y2 = np.cos(x1)
y3 = -np.cos(x1)
```

```python
plt.title("Line with Marker")

plt.plot(x1, y2, linestyle='dashed', color='m')

plt.plot(x1, y3, linestyle='dashdot', color='c')

plt.xlim([1.57, 4.72])
plt.ylim([-1.6, 1.6])
```

# 2. Scatter Plots:

### Create a basic scatter plot.

```python
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 10
colors = np.random.rand(50) * 10
area = (np.pi * (20 * np.random.rand(50))**2) 
```

```python
plt.scatter(x_scatter, y_scatter)
```

### Customize marker styles, sizes, and colors.

```python
plt.scatter(x_scatter, y_scatter, c=colors, cmap='Spectral', alpha=0.5, s = area)
```

### Plot multiple scatter plots on the same graph.

```python
plt.scatter(x_scatter, y_scatter, color='y', label='Scatter 1')
plt.scatter(np.arange(1,10), np.arange(1,10)+0, color='r', label='Scatter 2')
plt.show()
```

### Add labels and titles to the scatter plot.


```python
plt.scatter(x_scatter, y_scatter, color='m', label='Scatter 1')
plt.scatter((np.random.rand(50)*10), (np.random.rand(50)*10), color='g', label='Scatter 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Scatter Plots')
plt.show()
```

# 3. Bar Charts:

### Create a vertical bar chart.

```python
bars = ('A', 'B', 'C', 'D', 'E')
height = [3, 12, 5, 18, 45]
```

```python
plt.bar(bars, height)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Horizontal Bar Chart')
plt.show()
```

### Create a horizontal bar chart.

```python
plt.barh(bars, height)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Horizontal Bar Chart')
plt.show()
```

### Customize bar colors and add edge colors.

```python
plt.bar(bars, height, color='#00b894', edgecolor='red')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Customized Bar Chart')
plt.show()
```

# 4. Pie Charts:

```python
sizes = [15, 27, 30, 10, 22, 14, 24]
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
colors = ['blue', 'cyan', 'lightcoral', 'lightskyblue', 'magenta', 'pink', 'royalblue']
```

### Create a basic pie chart.

```python
plt.pie(sizes, labels=labels)
plt.title('Basic Pie Chart')
plt.show()
```

### Customize slice colors and add a legend.

```python
plt.pie(sizes, labels=labels, colors=colors)
plt.title('Pie Chart: Colors and Legend')
plt.show()
```

### Explode a slice of the pie chart.

```python
explode = (0.1, 0, 0.1, 0, 0, 0, 0)
```

```python
plt.pie(sizes, explode=explode, labels=labels, colors=colors)
plt.title('Exploded Pie Chart')
plt.show()
```

### Add percentage labels to slices.

```python
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart with Percentage Labels')
plt.show()
```
